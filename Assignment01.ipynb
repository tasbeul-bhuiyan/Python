## Exercise 16: Safe Pay Calculator

Task: Combine overtime + safe input:
- Ask for hours and rate
- If either is invalid: print `Error, please enter numeric input`
- Else compute overtime pay and print it


````python
# TODO: Write your solution here
h=input("Enter hours: ")
r=input("Enter rate: ")

if h.isnumeric() and r.isnumeric():
    h=float(h)
    r=float(r)
    if h>40:
        p=(40*r)+(h-40)*r*1.5
        print("Pay =",p)
    else:
        p=h*r
        print("Pay =",p)
else:
    print("Error,please enter neumeric input")
````

Task: Make a program that:

2. Takes hours + rate (numbers, safe input using try/except)
3. Computes overtime pay
4. Prints a clean payslip
