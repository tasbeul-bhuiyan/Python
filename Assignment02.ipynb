## Problem 24

Write a function `count_vowels(s)` that returns how many vowels are in `s` (a, e, i, o, u). Print the result for `'education'`.

```python
# TODO: Define count_vowels(s)

# TODO: Print count_vowels('education')
def count_vowels(s):
    count=0
    for letter in s:
        if letter in "aeiou":
            count=count+1
    return count
print(count_vowels("education"))
```

## Problem 25

Write a function `summarize(nums)` that returns a tuple:
`(count, total, average, smallest, largest)`
Constraints: no `len()`, `sum()`, `min()`, `max()`.
Test with `data = [12, 3, 19, 7, 7, 25]` and print the returned tuple.

```python
# TODO: Define summarize(nums)
# TODO: Compute count, total, smallest, largest using loop idioms
# TODO: Compute average after loop
# TODO: Return (count, total, average, smallest, largest)

data = [12, 3, 19, 7, 7, 25]
# TODO: Print summarize(data)
def summarize(nums):
    count=0
    total=0
    smallest=None
    largest=None
    for num in nums:
        total=total+num
        count=count+1
        if smallest==None or smallest>num:
            smallest=num
        if largest==None or largest<num:
            largest=num
    average=total/count
    tup=(count,total,average,smallest,largest)
    return tup
print(summarize(data))

```
